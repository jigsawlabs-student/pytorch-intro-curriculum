### Architecting our Network

$$
\begin{aligned}
z_1 & = xW_1 + b_1 \\
a_1 & = sigmoid(z_1) \\
z_2 & = a_1W_2 + b_2 \\
a_2 & = \hat{y} = \mathrm{softmax}(z_2)
\end{aligned}
$$

$softmax(ll_2(\sigma(ll_1(X))))$

```python
def sigmoid(x): return 1/(1 + np.exp(-x))
```

```python
sigmoid(0)
```

```python
def softmax(input_arr):
    raised = np.exp(input_arr)
    return raised / np.sum(raised, axis=1, keepdims = True)
```

$$
\begin{aligned}
z_1 & = xW_1 + b_1 \\
a_1 & = sigmoid(z_1) \\
z_2 & = a_1W_2 + b_2 \\
a_2 & = \hat{y} = \mathrm{softmax}(z_2)
\end{aligned}
$$

$softmax(ll_2(\sigma(ll_1(X))))$

```python
X.shape
```

```python
W_1 = np.random.randn(2,3)
b_1 = np.random.randn(3)
W_2 = np.random.randn(3, 2)
b_2 = np.random.randn(2)
```

```python
X[:3]
```

```python
(X@W_1 + b_1).shape
```

```python
def feed_forward(X):
    ll1 = (X@W_1 + b_1)
    L1 = sigmoid(ll1)
    ll2 = (L1@W_2 + b_2)
    y_hats = softmax(ll2)
    return y_hats
```
